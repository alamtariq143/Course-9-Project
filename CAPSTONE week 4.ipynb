### MY GOAL
I HAVE BEEN TASKED BY MY SUPERVISOR WITH THE RESPONSIBILITY TO PROVIDED SCIENTIFIC EVIDENCE USING THE DATA OF BOTH CITIES TO MAKE A PRESENTATION THAT SHOWCASE THE PREFERRED CITY TO START A RESTAURANT BUSINESS AND THE SPECIFIED NEIGHBORHOOD. THIS PRESENTATION SHOULD CONTAIN GEOGRAPHICAL IMAGES, TABLES OR DATA FRAMES, DATA OF RESTAURANTS IN NEIGHBORHOODS SITUATED IN BOTH CITIES EXTRACTED USING FOURSQUARE API.

### MY APPROACH
I RESEARCH THE VOLUME OF IMMIGRATION IN THE PAST 5 YEARS TO THE CITY OF NEW YORK AND TORONTO TO SEE WHICH OF THE CITY ATTRACTS MORE VISITORS ESPECIALLY FROM THE AFRICAN CONTINENT AND LOOK AT THE GENERAL RATE OF EATING OUT AND NIGHT LIFE IN BOTH CITIES. MY CLIENT'S GOAL IS TO COMPETE WITH TOP LOCAL RESTAURANTS DURING THE DAY AND PROVIDE AFRICAN ORIENTATED NIGHT LIFE FOR ALL.

HAVING CONSIDERED THE ABOVE FACTORS, I HAVE CHOSEN THE CITY OF NEW YORK FOR MY CLIENT AS IT HAS MORE VISITORS COMPARED WITH TORONTO AND NIGHT LIFE MIGHT BE HIGHLY APPRECIATED. WITH AN  INTERNATIONAL AIRPORT THAT BOAST TRAFFIC OF 61 MILLION PASSENGERS, IT IS THE BUSIEST AIRPORT IN THE NORTHERN AMERICA AND THE CENTER FOR INTERNATIONAL FLIGHT NETWORK.  NEW YORK CITY ATTRACTS VISITORS WITH THE PRESENCE OF FAMOUS ATTRACTIONS LIKE THE STATUE OF LIBERTY, CENTRAL PARK, BROADWAY AND THE THEATER DISTRICTS, VISITORS AND RESIDENCE ARE TREATED TO A VARIETY OF DAY AND NIGHT ENTERTAINMENT.


WORKING WITH PYTHON IN A JUPITER NOTEBOOK, I IMPORT ALL THE NECESSARY LIBRARY THAT I NEED TO RUN FUNCTIONS, THEN DOWNLOAD AND EXPLORE THE DATA OF NEW YORK CITY. THE RESULT IS A LIST OF NEIGHBORHOODS WITHIN NEW YORK, THEIR COORDINATES AND PROPERTIES. I CONVERT THIS LIST TO A PANDAS DATA FRAME FOR EASY VIEWING AND UNDERSTANDING THE DATA. THE DATA IS FILTERED TO ZOOM IN ON THE SPECIFIC NEIGHBORHOOD AS MY TARGET IN OTHER TO GENERATE SPECIFIC INFORMATION THAT RELATES TO THE BUSINESS OF MY CLIENT.

USING GEOPY LIBRARY I GENERATE THE COORDINATES FOR NEW YORK CITY  AND USE FOLIUM; A MAP GENERATING FUNCTION TO CREATE A MAP OF NEW YORK CITY AND INPOSE THE NEIGHBORHOOD ON THE MAP.I USE THE COORDINATES FOR NEIGHBORHOODS IN NEW YORK CITY  I MAKE CALLS ON FOURSQUARE API   TO GENERATE A LIST OF TOP 5 RESTAURANTS IN EACH NEIGHBORHOODS WITHIN 5KM RADIUS  AND THE TRAFFIC VOLUME OF PEOPLE IN AND OUT OF THE RESTAURANT AT THE TIME OF THE CALL. USING K-CLUSTERING; A FUNCTION THAT GROUPS VENUSES IN TERMS OF SIMILARITIES AND LOCATION I DETERMINE THE BEST POSSIBLE NEIGHBORHOOD AND LOCATION FOR MY CLIENTS TO INVEST AND OPEN HIS NEW RESTAURANT.
